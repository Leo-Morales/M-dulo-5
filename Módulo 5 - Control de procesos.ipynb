Importado de librerias necesarias:

```python
import pandas as pd
import pyspc as ps
from pyspc import *
```

Lectura del dataset "EKPC_hourly" desde el ordenador

```python
EKPC_data= pd.read_csv("EKPC_hourly.csv")
EKPC_data
```

Filtrado de las filas que contienen la cantidad de energía que se consume a horas 20:00 de cada día

```python
EKPC_20hours = EKPC_data[EKPC_data['Time'] == "20:00:00"]
EKPC_20hours.head()
```

Guardado de los datos en un archivo csv

```python
EKPC_20hours.to_csv("EKPC_20hours.csv")
EKPC_dat= pd.read_csv("EKPC_20hours.csv")
EKPC_dat.info()
```

Grafica con pyspc del promedio de la columna de consumo de energía en Mega Watts

```python
a = spc(EKPC_dat.EKPC_MW) + ewma()
print(a)
```

Grafica con pyspc del rango de la columna de consumo de energía en Mega Watts

```python
a = spc(EKPC_dat.EKPC_MW) + xmr()
print(a)
```
